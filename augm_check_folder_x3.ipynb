```python
import tensorflow as tf
import numpy as np
from keras.utils import np_utils
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import os
```

```python
train_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255,
                                                           rotation_range=15,
                                                           width_shift_range=0.2,
                                                           height_shift_range=0.1,
                                                           zoom_range=0.1)
```

```python
train_path = '/home/a/datasets/mnist_png_augm/training'
test_path = '/home/a/datasets/mnist_png_augm/testing'

training_set = train_gen.flow_from_directory(train_path,
                                             shuffle=False,
                                             batch_size=1,
                                             target_size=(28,28),
                                             color_mode='rgb')

test_set = train_gen.flow_from_directory(test_path,                                         
                                         shuffle=False,
                                         batch_size=1,
                                         target_size=(28,28),
                                         color_mode='rgb')
```

```python
def get_img(num):
    path = os.path.join(train_path, str(num))
    file = os.listdir(path)[0]
    return cv2.imread(os.path.join(path, file))
```

```python
def demo(training_set, num):
    plt.figure(figsize=(3,4))
    plt.imshow(get_img(num))
    plt.figure(figsize=(10,10))
    for i in range(9):
        img, lab = training_set[num]
        plt.subplot(3,3,i+1)
        plt.imshow(img[0])
        plt.title(np.argmax(lab[0]))
```

```python
num = 9
```

```python
demo(training_set, num)
```
