```python
import tensorflow as tf
import numpy as np
from keras.utils import np_utils
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import os
```

```python
train_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255,
                                                           rotation_range=15,
                                                           width_shift_range=0.2,
                                                           height_shift_range=0.1,
                                                           zoom_range=0.1)
```

```python
train_path = '/home/a/datasets/mnist_png/training/'
test_path = '/home/a/datasets/mnist_png/testing/'

training_set = train_gen.flow_from_directory(train_path,
                                             shuffle=False,
                                             batch_size=1,
                                             target_size=(28,28),
                                             color_mode='rgb')

test_set = train_gen.flow_from_directory(test_path,                                         
                                         shuffle=False,
                                         batch_size=1,
                                         target_size=(28,28),
                                         color_mode='rgb')
```

```python
def demo(train_dataset, digit, pos=0):
    filenames = training_set.filenames
    counter = 0
    for i in range(0,len(filenames)):
        if filenames[i][0] == str(digit):
            if counter != pos:
                counter += 1
            else:
                print(i)
                plt.figure(figsize=(4,3))
                plt.imshow(cv2.imread(training_set.filepaths[i]))
                plt.figure(figsize=(10,10))
                for j in range(9):
                    plt.subplot(3, 3 , j + 1)
                    plt.imshow(train_dataset[i][0][0])
                break
```
